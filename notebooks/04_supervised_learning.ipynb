```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import warnings
warnings.filterwarnings('ignore')
```

```python
X_train = pd.read_csv('X_train_PCA.csv')
y_train = pd.read_csv('y_train.csv')
X_test = pd.read_csv('X_test_PCA.csv')
y_test = pd.read_csv('y_test.csv')
```

```python
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
y_pred_LR = model_LR.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_LR))
```

```python
confusion_matrix(y_test, y_pred_LR)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, y_pred_LR), annot=True)
```

```python
model_SVM = SVC()
model_SVM.fit(X_train, y_train)
```

```python
y_pred_SVM = model_SVM.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_SVM))
```

```python
confusion_matrix(y_test, y_pred_SVM)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, y_pred_SVM), annot=True)
```

```python
model_DC = DecisionTreeClassifier()
model_DC.fit(X_train, y_train)
```

```python
y_pred_DC = model_DC.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_DC))
```

```python
confusion_matrix(y_test, y_pred_DC)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, y_pred_DC), annot=True)
```

```python
model_RFC = RandomForestClassifier()
model_RFC.fit(X_train, y_train)
```

```python
y_pred_RFC = model_RFC.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_RFC))
```

```python
confusion_matrix(y_test, y_pred_RFC)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(confusion_matrix(y_test, y_pred_RFC), annot=True)
```
